```python
#1. Import the necessary libraries.
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
numerical_df = pd.read_csv('numerical.csv')
categorical_df = pd.read_csv('categorical.csv')
categorical_df
```

```python
numerical_df.dtypes
```

```python
numerical_df.select_dtypes(include = object)
numerical_df=numerical_df.drop(['state','response','coverage','education','effective_to_date','employment_status','gender','location_code','marital_status','policy_type','policy','renew_offer_type','vehicle_class','vehicle_size'] , axis=1)
```

```python
numerical_df
```

```python
numerical_df=numerical_df.drop(['sales_channel'] , axis=1)
```

```python
categorical_df
```

```python
categorical_df.columns=pd.Series(categorical_df.columns).apply(lambda x: "_".join(y.lower() for y in x.split(" ")))
categorical_df=categorical_df.drop(['unnamed:_0'], axis=1)
categorical_df.head()
```

```python
categorical_df
```

```python
categorical_df['policy'] = categorical_df['policy'].str.replace('\w* ', '')
```

```python
categorical_df['education'] = np.where(categorical_df['education'].isin(['College','Bachelor']), 'Undergraduate', categorical_df['education'])
```

```python
categorical_df['employmentstatus'] = np.where(categorical_df['employmentstatus'].isin(['Medical Leave','Disabled','Retired']), 'Other', categorical_df['employmentstatus'])
```

```python
categorical_df['vehicle_class'] = np.where(categorical_df['vehicle_class'].isin(['Sports ','Luxury ']), 'Luxury Car', categorical_df['vehicle_class'])
```

```python
categorical_df = categorical_df.drop(['vehicle_size'], axis = 1)
```

```python
data= pd.merge(left = numerical_df,
                 right = categorical_df,
                 how = 'inner', 
                 left_on = "customer", 
                 right_on= "customer")

data.head()
```

```python
data.dtypes
```

```python
from datetime import datetime as dt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
```

```python
data['effective_to_date'] =  pd.to_datetime(data['effective_to_date'])
```

```python
y = data['total_claim_amount']
X = data.drop(['customer','total_claim_amount','effective_to_date'], axis=1)

# Creating a Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=45)
```

### Separate X_train and X_test into numerical and categorical (X_train_cat , X_train_num , X_test_cat , X_test_num)

```python
# Selecting the numeric columns
X_train_num = X_train.select_dtypes(include = np.number)
X_train_cat = X_train.select_dtypes(include = object)
X_test_num = X_test.select_dtypes(include = np.number)
X_test_cat = X_test.select_dtypes(include = object)
```

### Use X_train_num to fit scalers. Transform BOTH X_train_num and X_test_num.



```python
# Applying transformer on numeric Train data
transformer = MinMaxScaler().fit(X_train_num)
X_train_num_norm = transformer.transform(X_train_num)
X_test_num_norm = transformer.transform(X_test_num)

X_train_num_norm = pd.DataFrame(X_train_num_norm,columns=X_train_num.columns)
X_test_num_norm = pd.DataFrame(X_test_num_norm,columns=X_test_num.columns)

display(X_train_num_norm.head())
X_test_num_norm.head()
```

### Encode the categorical variables X_train_cat and X_test_cat (See the hint below for encoding categorical data!!!)



```python
X_train_cat
```

```python
X_test_cat
```

```python
X_train_cat_ohe=X_train_cat[['state','marital_status','policy_type','policy','renew_offer_type','sales_channel','vehicle_class']]
encoder = OneHotEncoder(drop='first').fit(X_train_cat_ohe)
encoded = encoder.transform(X_train_cat_ohe).toarray()
cols = encoder.get_feature_names_out(input_features=X_train_cat_ohe.columns)
X_train_cat_ohe = pd.DataFrame(encoded, columns=cols)
X_train_cat_ohe.head()
```

```python
X_train_cat["coverage"] = X_train_cat["coverage"].map({"Basic" : 0, "Extended" : 1, "Premium" : 2})
X_train_cat["employmentstatus"] = X_train_cat["employmentstatus"].map({"Employed" : 0, "Unemployed" : 1, "Other" : 2})
X_train_cat["location_code"] = X_train_cat["location_code"].map({"Suburban" : 0, "Rural" : 1, "Urban" : 2})

```

```python
X_train_cat_ord=X_train_cat[['coverage','employmentstatus','location_code']]
X_train_cat_ord = X_train_cat_ord.reset_index()
X_train_cat_ord = X_train_cat_ord.drop(['index'], axis=1)
X_train_cat_ord.head()
```

```python
X_train_processed = pd.concat([X_train_num_norm, X_train_cat_ohe, X_train_cat_ord], axis=1)
X_train_processed.head()
```

```python
display(X_train_processed.isna().sum())
display(X_train_processed.shape)
```

```python
X_test_cat
```

```python
X_test_cat_ohe=X_test_cat[['state','marital_status','policy_type','policy','renew_offer_type','sales_channel','vehicle_class']]
encoded = encoder.transform(X_test_cat_ohe).toarray()
cols = encoder.get_feature_names_out(input_features=X_test_cat_ohe.columns)
X_test_cat_ohe = pd.DataFrame(encoded, columns=cols)
X_test_cat_ohe.head()
```

```python
X_test_cat["coverage"] = X_test_cat["coverage"].map({"Basic" : 0, "Extended" : 1, "Premium" : 2})
X_test_cat["employmentstatus"] = X_test_cat["employmentstatus"].map({"Employed" : 0, "Unemployed" : 1, "Other" : 2})
X_test_cat["location_code"] = X_test_cat["location_code"].map({"Suburban" : 0, "Rural" : 1, "Urban" : 2})
```

```python
X_test_cat_ord=X_test_cat[['coverage','employmentstatus','location_code']]
X_test_cat_ord = X_test_cat_ord.reset_index()
X_test_cat_ord = X_test_cat_ord.drop(['index'], axis=1)
X_test_cat_ord.head()
```

```python
X_test_processed = pd.concat([X_test_num_norm, X_test_cat_ohe, X_test_cat_ord], axis=1)
X_test_processed.head()
```

```python
from sklearn import linear_model
from sklearn.neighbors import KNeighborsRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import r2_score

import warnings
warnings.filterwarnings('ignore')

lm = linear_model.LinearRegression()
lm.fit(X_train_processed,y_train)

predictions_test = lm.predict(X_test_processed)
r2=round(r2_score(y_test, predictions_test),2)
print("r2_score: ",r2)
```

```python
def best_model_decision(model_list,X_train,y_train,X_test,y_test):
    for model in model_list:
        #1. Linear Regression
        if model=='linear regression model':
            lm = linear_model.LinearRegression()
            lm.fit(X_train,y_train)
            print("Linear Regression:")
            print("r2_score (train): ",round(r2_score(y_train, lm.predict(X_train)),2))    
            print("r2_score (test): ",round(r2_score(y_test, lm.predict(X_test)),2))
    
        #2. KNeighborsRegressor
        elif model=='knn model':
            knn = KNeighborsRegressor()
            knn.fit(X_train,y_train)
            print("K Nearest Neighbours:")
            print("r2_score (train): ",round(r2_score(y_train, knn.predict(X_train)),2))    
            print("r2_score (test): ",round(r2_score(y_test, knn.predict(X_test)),2))
            
        elif model=='mlp':
            mlp = MLPRegressor(random_state=1, max_iter=500)
            mlp.fit(X_train,y_train)
            print("Multi-layer Perceptron:")
            print("r2_score (train): ",round(r2_score(y_train, mlp.predict(X_train)),2))    
            print("r2_score (test): ",round(r2_score(y_test, mlp.predict(X_test)),2))
```

```python
model_list=['linear regression model','knn','mlp']
best_model_decision(model_list,X_train_processed,y_train,X_test_processed,y_test)
```
